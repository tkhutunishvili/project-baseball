```python
import pandas as pd
from pybaseball import batting_stats_range
battingdata = batting_stats_range('2014-05-01', '2017-08-29')
battingdf = pd.DataFrame(battingdata)
steal = battingdf.groupby('Name')['SB'].sum()
stealdf = steal.to_frame()
stealdf.query('SB >=1')
stealdf.plot.bar()
```
