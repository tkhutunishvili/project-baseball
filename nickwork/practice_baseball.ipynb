```python
from pybaseball import batting_stats_range
```

```python
battingdata = batting_stats_range('2014-05-01', '2017-08-29')
battingdf = pd.DataFrame(battingdata)
```

```python
battingdf.columns
```

```python
steal = battingdf.groupby('Name')['SB'].sum()
stealdf = steal.to_frame()
```

```python
#again could do a month by month breakdown
#could look at a team perspective from year to year basis
stealdf.query('SB >=1')
stealdf.plot.bar()
```

```python
from pybaseball import pitching_stats_range
pitching = pitching_stats_range('2017-05-01', '2017-05-29')
pitchingdf = pd.DataFrame(pitching)
```

```python
pitchingdf.columns
```

```python
#could do a moving scatter plot tracking 
info = pitchingdf[['Name','IP', 'ERA','GS']]
starters = info.query('GS >= 1')
starters.plot.scatter('IP', 'ERA')
```
